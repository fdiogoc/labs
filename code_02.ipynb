```
imagem_1 = Image.open('/content/drive/My Drive/lab02/bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_2 = Image.open('/content/drive/My Drive/lab02/claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_3 = Image.open('/content/drive/My Drive/lab02/johannes-plenio-RwHv7LgeC7s-unsplash.jpg')


plt.figure(figsize=(20,4))

plt.subplot(131) #131 1 linha 3 colunas posição 1
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem_1)

plt.subplot(132) #132 1 linha 3 colunas posição 2
plt.title('Imagem 2')
plt.axis('off')
plt.axis('off')
plt.imshow(imagem_2)

plt.subplot(133) #133 1 linha 3 colunas posição 3
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem_3)

plt.show()
```

# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
imagem_1_arr = np.array(imagem_1)
imagem_2_arr = np.array(imagem_2)
imagem_3_arr = np.array(imagem_3)

red_1 = imagem_1_arr[:,:,0]
green_1 = imagem_1_arr[:,:,1]
blue_1 = imagem_1_arr[:,:,2]

red_2 = imagem_2_arr[:,:,0]
green_2 = imagem_2_arr[:,:,1]
blue_2 = imagem_2_arr[:,:,2]

red_3 = imagem_3_arr[:,:,0]
green_3 = imagem_3_arr[:,:,1]
blue_3 = imagem_3_arr[:,:,2]

plt.figure(figsize=(20,16))

plt.subplot(431) #131 1 linha 3 colunas posição 1
plt.title('Imagem 1')
plt.axis('off')
plt.imshow(imagem_1)

plt.subplot(432) #132 1 linha 3 colunas posição 2
plt.title('Imagem 2')
plt.axis('off')
plt.axis('off')
plt.imshow(imagem_2)

plt.subplot(433) #133 1 linha 3 colunas posição 3
plt.title('Imagem 3')
plt.axis('off')
plt.imshow(imagem_3)


##RED
plt.subplot(434) 
plt.hist(red_1.ravel(), color='red')
plt.ylim(top=8e6)

plt.subplot(435) 
plt.hist(red_2.ravel(), color='red')
plt.ylim(top=8e6)

plt.subplot(4,3,6) 
plt.hist(red_3.ravel(), color='red')
plt.ylim(top=8e6)

##GREEN
plt.subplot(437)
plt.hist(green_1.ravel(), color='green')
plt.ylim(top=8e6)

plt.subplot(438) 
plt.hist(green_2.ravel(), color='green')
plt.ylim(top=8e6)

plt.subplot(4,3,9) 
plt.hist(green_3.ravel(), color='green')
plt.ylim(top=8e6)


##BLUES
plt.subplot(4,3,10) 
plt.hist(blue_1.ravel(), color='blue')
plt.ylim(top=8e6)

plt.subplot(4,3,11) 
plt.hist(blue_2.ravel(), color='blue')
plt.ylim(top=8e6)

plt.subplot(4,3,12) 
plt.hist(blue_3.ravel(), color='blue')
plt.ylim(top=8e6)

plt.show()
```
